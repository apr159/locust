```python
import aeon
import pandas as pd
from aeon.datasets import load_from_tsfile
from aeon.transformations.collection.convolution_based import Rocket
from aeon.transformations.collection.feature_based import Catch22
from aeon.transformations.collection.feature_based._tsfresh import TSFreshFeatureExtractor
```

```python
X,y = load_from_tsfile('data/2022-dataset.ts')
```

```python
catch22 = Catch22(use_pycatch22 = True)
X_ = catch22.fit_transform(X)
df = pd.DataFrame(X_)
df['Resultado'] = y
df.to_csv('data/2022-catch22.csv')
```

```python
fr = TSFreshFeatureExtractor()
X_ = fr.fit_transform(X)
df = pd.DataFrame(X_)
df['Resultado'] = y
df.to_csv('data/2022-tsfresh.csv')
```
