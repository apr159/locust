```python
import pandas as pd
import numpy as np
import json

from sklearn.model_selection import cross_val_score
```

```python
source_dataset = pd.read_csv('dataset.csv')

source_dataset
```

```python
dataset = pd.DataFrame()
dataset['ndvi_1'] = source_dataset.apply(lambda row: np.mean(json.loads(row['ndvi'])[0:16]),axis=1)
dataset['ndvi_2'] = source_dataset.apply(lambda row: np.mean(json.loads(row['ndvi'])[17:32]),axis=1)
dataset['ndvi_3'] = source_dataset.apply(lambda row: np.mean(json.loads(row['ndvi'])[33:64]),axis=1)
dataset['ndvi_4'] = source_dataset.apply(lambda row: np.mean(json.loads(row['ndvi'])[65:80]),axis=1)
dataset['ndvi_5'] = source_dataset.apply(lambda row: np.mean(json.loads(row['ndvi'])[81:96]),axis=1)

dataset['LST_Day_1km_1'] = source_dataset.apply(lambda row: np.mean(json.loads(row['LST_Day_1km'])[0:16]),axis=1)
dataset['LST_Day_1km_2'] = source_dataset.apply(lambda row: np.mean(json.loads(row['LST_Day_1km'])[17:32]),axis=1)
dataset['LST_Day_1km_3'] = source_dataset.apply(lambda row: np.mean(json.loads(row['LST_Day_1km'])[33:64]),axis=1)
dataset['LST_Day_1km_4'] = source_dataset.apply(lambda row: np.mean(json.loads(row['LST_Day_1km'])[65:80]),axis=1)
dataset['LST_Day_1km_5'] = source_dataset.apply(lambda row: np.mean(json.loads(row['LST_Day_1km'])[81:96]),axis=1)

dataset['Lai_1'] = source_dataset.apply(lambda row: np.mean(json.loads(row['Lai'])[0:16]),axis=1)
dataset['Lai_2'] = source_dataset.apply(lambda row: np.mean(json.loads(row['Lai'])[17:32]),axis=1)
dataset['Lai_3'] = source_dataset.apply(lambda row: np.mean(json.loads(row['Lai'])[33:64]),axis=1)
dataset['Lai_4'] = source_dataset.apply(lambda row: np.mean(json.loads(row['Lai'])[65:80]),axis=1)
dataset['Lai_5'] = source_dataset.apply(lambda row: np.mean(json.loads(row['Lai'])[81:96]),axis=1)


dataset['Resultado'] = source_dataset['Resultado']
dataset.to_csv('dataset_2022.csv')
```

```python
from sklearn import linear_model, svm
from sklearn.calibration import cross_val_predict
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score,precision_score, recall_score
from sklearn.model_selection import StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb

dataset = pd.read_csv('dataset_2022.csv')
X = dataset.drop(['Resultado'], axis=1)
y = dataset['Resultado']

clf = svm.SVC(kernel='linear', C=1, random_state=42)
lm =  RandomForestClassifier()
skf = StratifiedKFold(n_splits=5)

y_ = cross_val_predict(lm, X, y, cv=skf)
conf_mat = confusion_matrix(y, y_)
print(accuracy_score(y, y_))
#print(precision_score(y, y_))
#print(recall_score(y, y_))

conf_mat
```
